# Case 3: Horse Weight Distribution

```python
plt.figure(figsize=(10, 6))
df['wgHr'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')

# Calculate statistics
mean_value = df['wgHr'].mean()
mode_value = df['wgHr'].median()
max_value = df['wgHr'].max()
min_value = df['wgHr'].min()
std_deviation = df['wgHr'].std()

# Plot vertical lines for mean, mode, max, min
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_value}')
plt.axvline(mode_value, color='green', linestyle='dashed', linewidth=1, label=f'Median: {mode_value}')
plt.axvline(max_value, color='orange', linestyle='dashed', linewidth=1, label=f'Max: {max_value}')
plt.axvline(min_value, color='purple', linestyle='dashed', linewidth=1, label=f'Min: {min_value}')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it contains columns 'ord' and 'sjS1fOrd'
# Create a pivot table where the rows are sjS1fOrd, columns are ord, and values are the count of occurrences
pivot_table = df6.pivot_table(index='sjS1fOrd', columns='ord', aggfunc='size', fill_value=0)

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt="d")
plt.title('Heat Map of Finishing Position vs Position After 200m')
plt.xlabel('Finishing Position')
plt.ylabel('Position After 200m')
plt.show()
```
